```python
df=df.rename(columns={"Employee ID":"ID","Title":"Job Title","BusinessUnit":"Business Unit","EmployeeStatus":"Status","EmployeeType":"Employment Type","EmployeeClassificationType":"Work Type","DepartmentType":
"Department","GenderCode":"Gender","RaceDesc":"Ethnicity","MaritalDesc":"Marital Status","Training Program Name":"Training Program"})
df
```

```python
df=df.drop(['PayZone'],axis=1)
```

```python
df = df.map(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
df.dtypes
```

```python
df[['StartDate','Survey Date','DOB','Training Date']]
```

```python
df[['Survey Date', 'DOB']]=df[['Survey Date', 'DOB']].apply(lambda x: x.str.replace('-', '/'))
df[['Survey Date', 'DOB']]
```

```python
print(df[['StartDate', 'Survey Date', 'DOB', 'Training Date']].head())
```

```python
df['StartDate'] = pd.to_datetime(df['StartDate'], format='%m/%d/%Y')
df['Survey Date'] = pd.to_datetime(df['Survey Date'], format='%d/%m/%Y')
df['DOB'] = pd.to_datetime(df['DOB'], format='%d/%m/%Y')
df['Training Date'] = pd.to_datetime(df['Training Date'], format='%m/%d/%Y')
df[['StartDate', 'Survey Date', 'DOB', 'Training Date']].dtypes
```

```python
df['Job Title'].unique()
```

```python
df['Job Title']=df['Job Title'].str.replace(r'.*Production Technician.*', 'Production Technician', regex=True)
df['Job Title']=df['Job Title'].str.replace(r'.*Network Engineer*', 'Network Engineer', regex=True)
df['Job Title']=df['Job Title'].str.replace(r'.*IT Manager.*', 'IT Manager', regex=True)
df['Job Title']=df['Job Title'].str.replace(r'.*Sr. DBA*', 'DBA', regex=True)
df['Job Title']=df['Job Title'].str.replace(r'.*Database Administrator*', 'DBA', regex=True)
df['Job Title']=df['Job Title'].str.replace(r'.*Data Architect.*', 'Data Architect', regex=True)
df['Job Title']=df['Job Title'].str.replace(r'.*BI Developer.*', 'BI Developer', regex=True)
df['Job Title'].unique()
```

```python
df['Division'].unique()
```

```python
df['Division']=df['Division'].str.replace(r'.*General.*', 'General', regex=True)
df['Division']=df['Division'].str.replace(r'.*Project Management.*', 'Project Management', regex=True)
df['Division'].unique()
```

```python
df['State'].unique()
```

```python
state = {
    'MA': 'Massachusetts', 'ND': 'North Dakota', 'FL': 'Florida', 'CT': 'Connecticut', 'CA': 'California',
    'OR': 'Oregon', 'TX': 'Texas', 'IN': 'Indiana', 'GA': 'Georgia', 'CO': 'Colorado', 'KY': 'Kentucky',
    'NV': 'Nevada', 'VT': 'Vermont', 'PA': 'Pennsylvania', 'RI': 'Rhode Island', 'VA': 'Virginia',
    'UT': 'Utah', 'ID': 'Idaho', 'MT': 'Montana', 'WA': 'Washington', 'AZ': 'Arizona', 'NH': 'New Hampshire',
    'AL': 'Alabama', 'OH': 'Ohio', 'NY': 'New York', 'ME': 'Maine', 'TN': 'Tennessee', 'NC': 'North Carolina'
}
df['State'] = df['State'].map(state)
df['State']
```

```python
#Visualization
```

```python
df['StartYear'] = df['StartDate'].dt.year
total_employees = df['StartYear'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
total_employees.plot(kind='bar',colormap='flare')
plt.title('Number of Employment per Year')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(4,4))
ax=sns.histplot(df['Age'],kde=True,color='purple')
plt.title('Age Distribution')
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10, 8))
gender_count=df.groupby('Gender')['Gender'].count()
marital_count=df.groupby('Marital Status')['Marital Status'].count()
ethnicity_count=df.groupby('Ethnicity')['Ethnicity'].count()

gender_count.plot.pie(autopct='%1.1f%%',colormap='flare',ax=axs[0,0])
marital_count.plot.pie(autopct='%1.1f%%',colormap='flare',ax=axs[0,1])
ethnicity_count.plot.pie(autopct='%1.1f%%',colormap='flare',ax=axs[1,0])
sns.countplot(x='State', data=df[df['State'].isin(df['State'].value_counts().nlargest(5).index)],palette='flare',ax=axs[1,1],hue='State',legend=False)
plt.ylabel('')
plt.xticks(rotation=45)
fig.suptitle("Employee Demographics")

plt.tight_layout()
plt.show()
```

```python
#Total Employees by Departmen/Divison/Job Ttitle
fig, axs = plt.subplots(3,1, figsize=(12, 9))
sns.countplot(x=df['Department'] , ax=axs[0],hue=df['Department'],palette='rocket')
sns.countplot(x='Division', data=df[df['Division'].isin(df['Division'].value_counts().nlargest(5).index)], ax=axs[1],hue=df['Division'],palette='rocket',legend=False)
sns.countplot(x='Job Title', data=df[df['Job Title'].isin(df['Job Title'].value_counts().nlargest(5).index)], ax=axs[2],hue=df['Job Title'],palette='rocket',legend=False)
plt.tight_layout()
plt.show()
```

```python
data=df[['Current Employee Rating', 'Engagement Score','Satisfaction Score', 'Work-Life Balance Score']].corr()
sns.heatmap(data,annot=True)
plt.title("Correlation Between Employee Ratings", fontsize=12)
plt.show()
```

```python
df['StartYear'] = df['StartDate'].dt.year
data = df.groupby('StartYear')[['Engagement Score', 'Satisfaction Score', 'Work-Life Balance Score']].mean()
plt.figure(figsize=(10, 6))
sns.lineplot(data=data,palette='rocket',markers=True)
plt.title('Average Scores Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Score')
plt.legend(title='Scores', labels=['Engagement', 'Satisfaction', 'Work-Life Balance'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Department', y='Training Outcome',hue=df['Department'],palette='rocket')
plt.title('Training Outcome by Department')
plt.xlabel('Department')
plt.ylabel('Training Outcome')
plt.xticks(rotation=45)
plt.show()
```

```python
outcome_counts = df.groupby('Training Program')['Training Outcome'].value_counts()

for program in outcome_counts.index.get_level_values(0).unique():
    program_data = outcome_counts[program]
    plt.figure(figsize=(4, 4))
    program_data.plot(kind='pie', autopct='%1.1f%%', legend=False,colormap='flare')
    
    plt.title(f'Training Outcome for {program}')
    plt.ylabel('')
    plt.show()
```

```python
avg_price=df.groupby('Training Program')['Training Cost'].mean().reset_index()
sns.lineplot(data=avg_price,x='Training Program',y='Training Cost',marker="o",color='purple')
plt.xticks(rotation=45)
plt.yticks([540,545,550,555,560,565,570]) 
plt.ylabel('Average Price')
plt.xlabel('Training Program')
plt.title('Average Price for Training Program')
plt.show()
```

```python
status = df.groupby(['Department', 'Status']).size().reset_index(name='Count')
status_pivot = status.pivot(index='Department', columns='Status', values='Count')

plt.figure(figsize=(10, 6))
sns.lineplot(data=status_pivot, markers=True)

plt.xticks(rotation=0)
plt.ylabel('Number of Employees')
plt.xlabel('Department')
plt.title('Employee Status by Department')
plt.show()
```
