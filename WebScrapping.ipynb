```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_cities_with_over_one_million_inhabitants'
```

```python
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.text,'html')
```

```python
soup.find('table',class_='static-row-numbers plainrowheaders vertical-align-top sticky-header sortable wikitable')
```

```python
table=soup.find('table',class_='static-row-numbers plainrowheaders vertical-align-top sticky-header sortable wikitable')
```

```python
titles=table.find_all('th')
```

```python
titles=[title.text.strip() for title in titles]
```

```python
titles
```

```python
import pandas as pd
```

```python
pd.DataFrame(columns=titles)
```

```python
df=pd.DataFrame(columns=titles)
df
```

```python
column=table.find_all('tr')
```

```python
for row in column[1:]:
    row_data=row.find_all('td')
    dataRow=[data.text.strip() for data in row_data]
    length=len(df)
    df.loc[length]=dataRow
```
